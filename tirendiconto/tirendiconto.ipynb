```python
from __future__ import absolute_import, division, print_function

import pandas as pd
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('tirendiconto.csv')
```

```python
df.loc[:, 'claimed_expenses'] /= 100
df.loc[:, 'total_expenses'] /= 100
df.loc[:, 'wire_transfer'] /= 100
```

```python
df.boxplot('claimed_expenses', by='month', figsize=(16, 9))
```

```python
df.boxplot('wire_transfer', by='month', figsize=(16, 9))
```

```python
df_claimed = pd.DataFrame({column: values['claimed_expenses'] for column, values in df.groupby('name')})
medians = df_claimed.median().sort_values()
df_claimed[medians.index].boxplot(figsize=(16, 36), vert=False)
```

```python
df_wire = pd.DataFrame({column: values['wire_transfer'] for column, values in df.groupby('name')})
medians = df_wire.median().sort_values(ascending=False)
df_wire[medians.index].boxplot(figsize=(16, 36), vert=False)
```
